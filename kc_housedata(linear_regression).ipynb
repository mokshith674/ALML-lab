```python
import warnings
warnings.filterwarnings('ignore')
df['year']=pd.DatetimeIndex(df['date']).year
df['month']=pd.DatetimeIndex(df['date']).month
df['day']=pd.DatetimeIndex(df['date']).day
df
```

```python
df.drop('date',axis=1,inplace=True)
df
```

```python
cat_features=['waterfront', 'view', 'condition', 'grade','year','month','day','zipcode']
df=pd.get_dummies(df,columns=cat_features)
print(df.columns)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_normalized=scaler.fit_transform(df[['price', 'bedrooms', 'bathrooms', 'sqft_living','floors']])
df_normalized
```

```python
normalize=pd.DataFrame(df_normalized,columns=['price', 'bedrooms', 'bathrooms', 'sqft_living','floors'])
normalize=normalize.join(df[df.columns.drop(['price', 'bedrooms', 'bathrooms', 'sqft_living','floors'])])
```

```python
normalize
```

```python
x=normalize.drop('price',axis=1)
x
```

```python
y=normalize['price']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=100)
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
y_pred=model.predict(xtest)
y_pred
```

```python
print('train_accuracy:',model.score(xtrain,ytrain))
print('test_accuracy:',model.score(xtest,ytest))
```

```python
x_pred=model.predict(xtrain)
x_pred
```

```python
from sklearn.metrics import mean_squared_error 
train_RMSE=mean_squared_error(ytrain,x_pred)**0.5
print('train_RMSE:',train_RMSE)
train_RMSE=mean_squared_error(ytest,y_pred)**0.5
print('test_RMSE:',train_RMSE)
```
