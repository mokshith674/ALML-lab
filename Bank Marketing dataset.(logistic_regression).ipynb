# Consider duration, age, and campaign columns as predictors and the column “y”(states whether the client will subscribe a term deposit or not) as target  variable

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("D:\\Dataset\\bank-additional-full.csv")
df
```

```python
x=df.drop(['job','marital','education','default','housing','loan','contact','month','day_of_week','pdays','previous','poutcome',
          'emp.var.rate','cons.price.idx','euribor3m','nr.employed','y','cons.conf.idx'],axis=1)
x
```

```python
y=df['y']
y
```

```python
#Use 70% of the data as training data set and 30% of data as testing data set.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Build a logistic regression model.  
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
train_prediction=model.predict(x_train)
train_prediction
```

```python
test_prediction=model.predict(x_test)
test_prediction
```

```python
#Determine the classification accuracy rate of the model on both train and test set.
train_accuracy=model.score(x_train,y_train)
print('train_acuracy:',train_accuracy)
test_accuracy=model.score(x_test,y_test)
print('test_acuracy:',test_accuracy)
```
