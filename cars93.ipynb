```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/Cars931611042729436.zip")
df
```

```python
columns = ["Manufacturer","Model","Type","Price","MPG.city","MPG.highway","Horsepower","Rear.seat.room","Passengers"]
df[columns].head()
```

```python
df["Price"].plot(kind="box",figsize = (5,5))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_figwidth(5)
fig.set_figheight(5)
ax1.boxplot(df["Horsepower"])
ax2.boxplot(df["MPG.city"])
ax1.set_title('Horsepower')
ax1.set_xlabel('Horsepower')
ax2.set_title('City Mileage')
ax2.set_xlabel("City Mileage (in miles per US gallon)")
fig.tight_layout()
```

```python
fig, ax = plt.subplots(2, 3)
fig.set_figwidth(10)
fig.set_figheight(7)
fig.suptitle("Multiple Box Plots", fontsize=16)
ax[0][0].boxplot(df["Price"][df["Type"]=="Compact"])
ax[0][0].set_title('Compact')
ax[0][1].boxplot(df["Price"][df["Type"]=="Large"])
ax[0][1].set_title('Large')
ax[0][2].boxplot(df["Price"][df["Type"]=="Midsize"])
ax[0][2].set_title('Midsize')
ax[1][0].boxplot(df["Price"][df["Type"]=="Small"])
ax[1][0].set_title('Small')
ax[1][1].boxplot(df["Price"][df["Type"]=="Sporty"])
ax[1][1].set_title('Sporty')
ax[1][2].boxplot(df["Price"][df["Type"]=="Van"])
ax[1][2].set_title('Van')
```

```python
ax = df.plot(["Horsepower"],["MPG.city"],kind="scatter", color = "black",marker = "*",figsize=(5,5))
ax.set_xlabel("Horsepower")
ax.set_ylabel("MPG.city")
ax.set_title("Horsepower vs MPG.city",fontsize=16)
```

```python
fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(7)
car_type_list = df["Type"].unique()
print(car_type_list)
colors_list = ['red','blue','pink','green','black','yellow']
for car_type,colr in zip(car_type_list,colors_list):
  x = df[df["Type"] == car_type]["Horsepower"]
  y = df[df["Type"] == car_type]["MPG.city"]
  plt.scatter(x,y,color = colr,label=car_type)
plt.suptitle("Scatter plot of horsepower and mileage",fontsize=16)
plt.xlabel("Horsepower")
plt.ylabel("Mileage City")
plt.legend()
```

```python
fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(7)
plt.bar(df["DriveTrain"], df["MPG.city"],width=0.2,label="Mileage in city")
plt.suptitle("DriveTrain vs MPG.city",fontsize=16)
plt.xlabel("DriveTrain")
plt.ylabel("MPG.city")
plt.legend()
```

```python
plt.barh(df["DriveTrain"], df["MPG.city"],height=0.2,label="Mileage in city")
plt.suptitle("DriveTrain vs MPG.city",fontsize=16)
plt.xlabel("DriveTrain")
plt.ylabel("MPG.city")
plt.legend()
```
